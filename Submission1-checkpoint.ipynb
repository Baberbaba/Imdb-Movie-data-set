## Now we want the number of  Top-rated movies having `Pg-13` rating produced in `USA`

```python
# Sorting the DataFrame as per the rating and assigning it to variable
sorted_movies = detailed_movies.sort_values(by='IMDb_rating',ascending=False)
sorted_movies
```

## Now for the sorted DataFrame we need to filter as per country and content rating

```python
# using the loc function to filter the columns based on each other
filtered_movies = sorted_movies.loc[sorted_movies['Country']=='USA'].loc[sorted_movies['content_rating']=='PG-13']
# Checking the count and assigning it to variable 
number_of_movies = filtered_movies['Title'].count()
print(f'Number of the top-rated movies produced in the United States having PG-13 rating is {number_of_movies}')
```

### We sorted the movies in descending order of _imdb rating_ and later filtered it based on _Country_ and _content rating_

```python
# filtering again using the loc function and assigning it to a variable 
non_us_movies = sorted_movies.loc[sorted_movies['title_year']==2014].loc[sorted_movies['Country']!='USA']
non_us_movies
```

```python
# checking the count of non us movies and assigning it to variable
number_non_us = non_us_movies['Title'].count()
```

```python
# checking the  non_us_movies and creating a series based on it
# since in number_non_us will have index from 0 we need to specify our range from 1-4
title_non_us = pd.Series(non_us_movies['Title'].values,index=(np.arange(1,number_non_us+1)))
```

```python
# Printing the number of movies and the movies in series
print(f'Top-rated movies of 2014 which were were not produced in the United States are:\n{title_non_us}')
```

```python
#Creating new variable with all values of genres
combined_genre=sorted_movies.loc[:,'genre_1':'genre_3']
# creating a new column and using lamda function . Since NaN != NaN
sorted_movies['genree']=combined_genre.applymap(lambda value : 1 if value == value else 0).sum(axis=1)

sorted_movies
```

```python
# Now count the occurence of 3 ,2 ,1 ie movies with 1 genre only , 2 genres only , 3 genres only
one_genre_only = sorted_movies.loc[sorted_movies['genree']==1,'Title'].count()
two_genres_only = sorted_movies.loc[sorted_movies['genree']==2,'Title'].count()
three_genres_only = sorted_movies.loc[sorted_movies['genree']==3,'Title'].count()
```

```python
# Counting total movies
total_movies = sorted_movies['Title'].count()
```

```python
# Calculating Percentage based on genres and rounded of the deciaml to 0 decimal places
print(f'The percentage of the top-rated movies that are in 1 genre only is {(one_genre_only/total_movies)*100:.0f}%')
print(f'The percentage of the top-rated movies that are in 2 genres only is {(two_genres_only/total_movies)*100:.0f}%')
print(f'The percentage of the top-rated movies that are in 3 genres only is {(three_genres_only/total_movies)*100:.0f}%')
```

```python
# creating a list using list comprehension for desired column.
list1 = [x for x in detailed_movies['actor_1_name']]
```

```python
list2 = [x for x in detailed_movies['actor_2_name']]
```

```python
list3 = [x for x in detailed_movies['actor_3_name']]
# removing _'0'_ from list 3 
list3.remove('0')
```

```python
# concatinating lists into 1 list
list4 = list1 + list2 + list3
```

```python
# Creating a new list of alphabetically sorted values
actors=sorted(list4)
```

```python
# removing duplicates and appending the values in a empty list
top_actors=[]
for values in actors:
    if values not in top_actors:
        top_actors.append(values)
print(top_actors)
```
